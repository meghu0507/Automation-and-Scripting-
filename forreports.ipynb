```python
import os
import numpy as np 
import pandas as pd
os.getcwd()

files = os.listdir("Modules")

files = [file for file in files if ".v" in file]
##print(files)

top_modules = [file for file in files if "top" in file]
with open("Impl/top_CLA8/timing.txt") as f:
    lines = f.read()   
#print(lines)

timing = lines.split("\n") 
#print(timing)

timing_info = [line for line in timing if "Data Path Delay" in line]
##print(timing_info) 

dpd = float(timing_info[0].split()[3][:-2])
#print(dpd) 

DatapathDelay = []
#print(DatapathDelay)
DynamicPower = []
Slice_LUT = []
F7_Muxes = [] 
StaticPower = [] 
LogicPower = []
SignalPower = [] 
TotalPower = [] 
LUT_As_Memory = [] 
LUT_As_Logic = [] 
LUT_2 = [] 
LUT_4 = [] 



for mod in top_modules[:-1]:
    with open("Impl/{}/timing.txt".format(mod[:-2])) as f:
        lines = f.read()
        timing = lines.split("\n")
        timing_info = [line for line in timing if "Data Path Delay" in line]
        dpd = float(timing_info[0].split()[3][:-2])
        DatapathDelay.append(dpd)
        
        
    with open("Impl/{}/power.txt".format(mod[:-2])) as f:
        lines = f.read()
        power = lines.split("\n")
        power_info = [line for line in power if "Dynamic" in line]
        dpower = float(power_info[0].split("|")[2].strip())
        DynamicPower.append(dpower)
        power_info1 = [line for line in power if "Device Static" in line]
        spower = float(power_info1[0].split("|")[2].strip())
        StaticPower.append(spower)
        power_info2 = [line for line in power if "Slice Logic" in line]
        lpower = float(power_info2[0].split("|")[2].strip())
        LogicPower.append(lpower)
        power_info3 = [line for line in power if "Signals" in line]
        sigpower = float(power_info3[0].split("|")[2].strip())
        SignalPower.append(sigpower)
        
        power_info4 = [line for line in power if "Total On-Chip Power" in line]
        print (power_info4[0].split("|")[2].strip())
        #if "Total On-Chip Power" in line 
        totalpower = (power_info4[0].split("|")[2].strip())
        #print (power_info4)
        if len(totalpower) > 5:
            TotalPower.append(totalpower.split("(")[0].strip())
        else:
            TotalPower.append(totalpower)
        
        
    with open("Impl/{}/utilization.txt".format(mod[:-2])) as f:
        lines = f.read()
        uti = lines.split("\n")
        uti_info = [line for line in uti if "Slice LUTs" in line]
        util = int(uti_info[0].split("|")[2].strip())
        Slice_LUT.append(util)   
        
        uti_info1 = [line for line in uti if "F7 Muxes" in line]
        util_1 = int(uti_info1[0].split("|")[2].strip())
        F7_Muxes.append(util_1)
        
        uti_info2 = [line for line in uti if "LUT as Memory" in line]
        util_2 = int(uti_info2[0].split("|")[2].strip())
        LUT_As_Memory.append(util_2)
        
        uti_info3 = [line for line in uti if "LUT as Logic" in line]
        util_3 = int(uti_info3[0].split("|")[2].strip())
        LUT_As_Logic.append(util_3)
        
        try: 
            uti_info4 = [line for line in uti if "LUT2" in line]
            util_4 = (uti_info4[0].split("|")[2])
            
            LUT_2.append(util_4.strip())
        except: 
            print("Exception Handle") 
            LUT_2.append("NA") 
            
        
        try: 
            uti_info5 = [line for line in uti if "LUT4" in line]
            util_5 = (uti_info5[0].split("|")[2])
            
            LUT_4.append(util_5.strip())
        except: 
            print("Exception Handle") 
            LUT_4.append("NA") 
            
        
        
        
#print(DatapathDelay)
#print(DynamicPower)
#print(Utilization)
    
result = [DatapathDelay, DynamicPower, StaticPower,LogicPower,SignalPower,TotalPower, Slice_LUT, F7_Muxes,LUT_As_Logic, LUT_As_Memory, LUT_2, LUT_4 ]
resultT = np.array(result,dtype = object ).T
heading = ["DataPath Delay", "DynamicPower","StaticPower","LogicPower","Signal Power","TotalPower", "Slice_LUT", "F7_Muxes","LUT_As_Logic", "LUT_As_Memory", "LUT_2", "LUT_4"]
df = pd.DataFrame(resultT)     
df.columns = heading      
#print(result)
#print(resultT)
#print(heading) 

df.insert(loc=0, column='Design', value=top_modules[:-1]) 
print(df)
df.to_csv("result.csv")
print(df) 
```
